```python
from statsbombpy import sb
import pandas as pd
```

```python
comps = sb.competitions()
comps
```

```python
matches = sb.matches(competition_id=9, season_id=27)
matches
```

```python
matches_bremen = matches[(matches["home_team"] == "Werder Bremen") | (matches["away_team"] == "Werder Bremen")]
matches_bremen
```

```python
shots_junuzovic = pd.DataFrame()
```

```python
for id in matches_bremen["match_id"]:
    events_now = sb.events(match_id = id)
    shots = events_now[(events_now["player"] == "Zlatko Junuzović") & (events_now["type"] == "Shot")]
    shots_junuzovic = shots_junuzovic.append(shots)
```

```python
shots_junuzovic.head()
```

```python
shots_junuzovic.columns
```

```python
columns_to_keep = ['shot_statsbomb_xg','shot_outcome','shot_type','location']
shots_df = shots_junuzovic.loc[:,columns_to_keep]
```

```python
shots_df
```

```python
goals = shots_df[shots_df["shot_outcome"] == "Goal"]
goals
```

```python
shots_df[["x","y"]] = shots_df["location"].apply(pd.Series)
```

```python
shots_df.head()
```

```python
import matplotlib.pyplot as plt
from highlight_text import fig_text
import matplotlib as mpl
from mplsoccer import Pitch, VerticalPitch
import matplotlib.image as mpimg
import matplotlib.colors as mcolors
```

```python
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

background_img = plt.imread('werder_logo.png')


fig, ax = plt.subplots(figsize=(13, 8.5))
fig.set_facecolor('#82a67d')
ax.patch.set_facecolor('#82a67d')

pitch = VerticalPitch(pitch_type='statsbomb', half=True,
                      pitch_color='#82a67d', line_color='#c7d5cc')
pitch.draw(ax=ax)

ax.imshow(background_img, extent= [6.0, 74.0,66.0, 114.0], aspect='auto', alpha = 0.15)

for index, shot in shots_df.iterrows():
    if shot["shot_outcome"] == "Goal":
        plt.scatter(shot['y'], shot['x'], s=500*shot["shot_statsbomb_xg"], c='#0a481e', alpha=.7)
    else:
        plt.scatter(shot['y'], shot['x'], s=500*shot["shot_statsbomb_xg"], c='white', alpha=.7)

text_color = 'w'
total_shots = len(shots_df)
xG = round(sum(shots_df["shot_statsbomb_xg"]), 2)
plt.figtext(s=f'Total Shots: {total_shots}', x=.27, y=.17, fontsize=14, color=text_color)
plt.figtext(s=f'xG: {xG}', x=.49, y=.17, fontsize=14, color=text_color)
plt.figtext(s=f'Goals: 4', x=.68, y=.17, fontsize=14, color=text_color)
ax.set_title('All Shots of Zlatko Junuzović Season 15/16 ', fontsize=16, fontweight='bold', color= text_color)

fig.text(.165,.13,'plot point size representative of XG',fontstyle='italic',fontsize=11,color=text_color)
fig.text(.4,.13,'dark circles represent goals',fontstyle='italic',fontsize=11,color=text_color)
fig.text(.7,.13,'data from statsbomb',fontstyle='italic',fontsize=11,color=text_color)
plt.show()
```
